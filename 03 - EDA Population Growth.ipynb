```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load dataset
file_path = 'data/Data Sensus 2022.xlsx'
data_population = pd.read_excel(file_path, sheet_name=0)
data_population.head()
```

```python
# Checking data information
data_population.info()
```

### Deskriptif Analysis Penduduk

```python
# Urutkan data berdasarkan jumlah Populasi
data_sorted = data_population.sort_values(by='Population', ascending=False)

# Ambil 5 provinsi dengan populasi tertinggi
top_5_provinces = data_sorted.head(5)

# Hitung total jumlah laki-laki dari 5 provinsi teratas
total_population_top_5 = top_5_provinces['Population'].sum()

# Hitung persentase untuk 5 provinsi teratas
top_5_provinces['Persentase'] = (top_5_provinces['Population'] / total_population_top_5) * 100

# Buat plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Persentase', y='Provinsi', data=top_5_provinces, ci=None, palette='viridis')

# Menambahkan keterangan angka pada setiap batang
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 1, p.get_y() + p.get_height() / 2, f'{width:.1f}%', 
            ha='left', va='center', fontsize=10, color='black')

# Label dan judul
plt.xlabel('Persentase dari Total Penduduk (%)')
plt.ylabel('Provinsi')
plt.title('5 Provinsi dengan Populasi Tertinggi di Indonesia')


plt.show()
```

```python
# Urutkan data berdasarkan jumlah Kematian dari yang terbesar ke terkecil
data_sorted = data_population.sort_values(by='Kematian', ascending=False)

# Ambil 5 provinsi dengan tingkat kematian tertinggi
top_5_provinces = data_sorted.head(5)

# Hitung total jumlah Kematian dari 5 provinsi teratas
total_kematian_top_5 = top_5_provinces['Kematian'].sum()

# Hitung persentase untuk 5 provinsi teratas
top_5_provinces['Persentase'] = (top_5_provinces['Kematian'] / total_kematian_top_5) * 100

# Buat plot batang
plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Persentase', y='Provinsi', data=top_5_provinces, ci=None, palette='viridis')

# Menambahkan keterangan angka pada setiap batang
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 1, p.get_y() + p.get_height() / 2, f'{width:.1f}%', 
            ha='left', va='center', fontsize=10, color='black')

# Label dan judul
plt.xlabel('Persentase dari Total Kematian Penduduk (%)')
plt.ylabel('Provinsi')
plt.title('5 Provinsi dengan Jumlah Kematian Tertinggi di Indonesia')

plt.show()
```

```python
# Urutkan data berdasarkan jumlah laki-laki dari yang terbesar ke terkecil
data_sorted = data_population.sort_values(by='Laki-laki', ascending=False)

# Hitung total jumlah laki-laki
total_laki_laki = data_sorted['Laki-laki'].sum()

# Hitung persentase untuk setiap provinsi
data_sorted['Persentase'] = (data_sorted['Laki-laki'] / total_laki_laki) * 100

# Buat plot batang
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='Persentase', y='Provinsi', data=data_sorted, ci=None, palette='viridis')

# Menambahkan keterangan angka pada setiap batang
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 1, p.get_y() + p.get_height() / 2, f'{width:.1f}%', 
            ha='left', va='center', fontsize=10, color='black')

# Menambahkan garis batas di nilai rata-rata persentase
average_percentage = data_sorted['Persentase'].mean()
ax.axvline(x=average_percentage, color='red', linestyle='--', label=f'Rata-rata: {average_percentage:.1f}%')
ax.legend()

# Label dan judul
plt.xlabel('Persentase dari Total Penduduk Laki-laki (%)')
plt.ylabel('Provinsi')
plt.title('Frekuensi Penduduk Laki-laki di Indonesia Berdasarkan Provinsi')

# Putar label sumbu X
plt.xticks(rotation=45)

plt.show()
```

```python
# Urutkan data berdasarkan jumlah perempuan dari yang terbesar ke terkecil
data_sorted = data_population.sort_values(by='Perempuan', ascending=False)

# Hitung total jumlah perempuan
total_perempuan = data_sorted['Perempuan'].sum()

# Hitung persentase untuk setiap provinsi
data_sorted['Persentase'] = (data_sorted['Perempuan'] / total_perempuan) * 100

# Buat plot batang
plt.figure(figsize=(12, 8))
ax = sns.barplot(x='Persentase', y='Provinsi', data=data_sorted, ci=None, palette='viridis')

# Menambahkan keterangan angka pada setiap batang
for p in ax.patches:
    width = p.get_width()
    ax.text(width + 1, p.get_y() + p.get_height() / 2, f'{width:.1f}%', 
            ha='left', va='center', fontsize=10, color='black')

# Menambahkan garis batas di nilai rata-rata persentase
average_percentage = data_sorted['Persentase'].mean()
ax.axvline(x=average_percentage, color='red', linestyle='--', label=f'Rata-rata: {average_percentage:.1f}%')
ax.legend()

# Label dan judul
plt.xlabel('Persentase dari Total Penduduk Perempuan (%)')
plt.ylabel('Provinsi')
plt.title('Frekuensi Penduduk Perempuan di Indonesia berdasarkan Provinsi')

# Putar label sumbu X
plt.xticks(rotation=45)

plt.show()
```

```python
# Menghitung selisih antara 'Laki-laki' dan 'Perempuan'
data_population['Selisih'] = data_population['Laki-laki'] - data_population['Perempuan']

# Membulatkan hasil selisih ke bilangan bulat terdekat
data_population['Selisih'] = data_population['Selisih'].round()

# Mengubah format kolom 'Selisih' menjadi string dengan koma sebagai pemisah desimal
data_population['Selisih'] = data_population['Selisih'].apply(lambda x: f"{int(x):,}".replace(',', '.'))
data_population = data_population.sort_values(by='Population', ascending=False)
data_population.head()
```

```python
# Persentase pengurangan populasi
persentase_pengurangan = 10  # Misalnya 10%

# Menghitung populasi setelah pengurangan
data_population['Populasi Sesudah'] = data_population['Population'] * (1 - persentase_pengurangan / 100)

# Menghitung frekuensi atau tingkat kematian
data_population['Tingkat Kematian'] = data_population['Kematian'] / data_population['Populasi Sesudah']

# Mengubah format 'Tingkat_Kematian' menjadi persentase dan membulatkannya
data_population['Tingkat Kematian (%)'] = (data_population['Tingkat Kematian'] * 100).round(2)

# Menampilkan DataFrame akhir
data_population[['Provinsi','Population', 'Kematian', 'Populasi Sesudah', 'Tingkat Kematian (%)']]
```

### Time Series Populasi Penduduk

```python
file_path = 'data/Data Sensus 2022.xlsx'
dts_population = pd.read_excel(file_path, sheet_name=6)
```

```python
dts_population.info()
```

```python
# Mengubah tipe data kolom menjadi integer
dts_population[2022] = dts_population[2022].astype(int)
dts_population[2023] = dts_population[2023].astype(int)
dts_population[2024] = dts_population[2024].astype(int)
```

```python
dts_population.head()
```

```python
# Hitung total untuk setiap kategori
dts_goal = dts_population.copy()
totals = {
    'Tahun': [
        '2022',
        '2023',
        '2024'
    ],
    'Total Penduduk': [
        dts_goal[2022].sum(),
        dts_goal[2023].sum(),
        dts_goal[2024].sum()
    ]
}

# Membuat DataFrame dari dictionary totals
df_totals = pd.DataFrame(totals)

# Menampilkan tabel
df_totals
```

```python
# Menghitung pertumbuhan populasi tahunan
dts_population.set_index('Provinsi', inplace=True)
df_growth = dts_population.pct_change(axis=1) * 100 

# Membulatkan hasil pertumbuhan ke dua tempat desimal
df_growth = df_growth.round(2)

# Mengganti NaN dengan 0 jika diperlukan
df_growth = df_growth.fillna(0)
```

```python
# Menambahkan tahun-tahun sebagai nama kolom untuk hasil pertumbuhan
df_growth.columns = [f'Pertumbuhan_{col}' for col in df_growth.columns]

# Mengurutkan berdasarkan pertumbuhan 2023 dan 2024 dari yang terbesar
df_growth_sorted = df_growth.sort_values(by=['Pertumbuhan_2023', 'Pertumbuhan_2024'], ascending=[False, False])

# Menampilkan hasil
df_growth_sorted
```
